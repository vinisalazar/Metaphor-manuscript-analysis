```python
gs_kw = dict(width_ratios=[1, 1], height_ratios=[1, 1])
    
fig, axd = plt.subplot_mosaic([['upper left', 'upper right'],
                               ['lower left', 'lower right']],
                              figsize=(15, 12), gridspec_kw=gs_kw, layout="constrained",)
                              # subplotpars=mpl.figure.SubplotParams(top=0, bottom=-0.01, right=0, left=-0.01, hspace=0, wspace=0))

im1 = "./output/postprocessing/marmg_sasb/runtime_sum.png"
im2 = "./output/postprocessing/marmg_sasb/runtime_sample.png"
im3 = "./output/postprocessing/marmg_sasb/memory_sum.png"
im4 = "./output/postprocessing/marmg_sasb/memory_sample.png"
    
for k, im, letter in zip(axd, [im1, im2, im3, im4], "ABCD"):
    im = mpimg.imread(im)
    axd[k].imshow(im, origin="lower",)# vmin=vmin, vmax=vmax)
    axd[k].invert_yaxis()
    axd[k].axes.get_xaxis().set_visible(False)
    axd[k].axes.get_yaxis().set_visible(False)
    axd[k].set_frame_on(False)
    axd[k].annotate(letter, xy=(0.2, 0.95), xycoords="axes fraction", fontsize=12, weight="bold")


plt.savefig("../figures/Fig4.png", dpi=600, bbox_inches="tight", pil_kwargs={"compression": "tiff_lzw"})
```
