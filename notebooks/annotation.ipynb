```python
gs_kw = dict(width_ratios=[1.1, 1], height_ratios=[1, 1])
    
fig, axd = plt.subplot_mosaic([['left', 'right'],
                               ['left', 'right']],
                              figsize=(15, 15), gridspec_kw=gs_kw, layout="constrained",)
                              # subplotpars=mpl.figure.SubplotParams(top=0, bottom=-0.01, right=0, left=-0.01, hspace=0, wspace=0))

im1 = "../annotation/strmg_marmg_categories_heatmap.png"
im2 = "../annotation/marmg_cacb/cog/plots/concatenated_phylum_barplot.png"
    
for k, im, letter in zip(axd, [im1, im2], "AB"):
    im = mpimg.imread(im)
    axd[k].imshow(im, origin="lower",)# vmin=vmin, vmax=vmax)
    axd[k].invert_yaxis()
    axd[k].axes.get_xaxis().set_visible(False)
    axd[k].axes.get_yaxis().set_visible(False)
    axd[k].set_frame_on(False)

    
axd["left"].annotate("A", xy=(0.1, 1.0), xycoords="axes fraction", fontsize=12, weight="bold")
_ = axd["right"].annotate("B", xy=(0.0, 0.975), xycoords="axes fraction", fontsize=12, weight="bold")

plt.savefig("../figures/Fig3.png", dpi=600, bbox_inches="tight")#, pil_kwargs={"compression": "tiff_lzw"})
```
